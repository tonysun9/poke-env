```python
def get_report(df):
    f_df = df[~((df['turns'] < 1) | ((df['win_method'] == 11) & (df['opp_pkmn_left'] == 6)))]
    # df_filtered = df[(df['turns'] < 1) | (df['win_method'] == '11' & df['opp_pkmn_left'] == 6)]
    print("Total battles: ", len(f_df))
    print("Win rate: ", float(len(f_df[f_df['result'] == 1]))/len(f_df))
    print("Average turns: ", f_df['turns'].mean())
    print("Average Pokemon left: ", f_df['pkmn_left'].mean())
    print("Average Opponent Pokemon left: ", f_df['opp_pkmn_left'].mean())
    print("Average Opponent Elo: ", f_df['opp_elo'].mean())
    return f_df
    
```

```python
f_r_df = get_report(r_df)
f_r_df
```
