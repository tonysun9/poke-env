```python
import json

with open("gen1randbats.json", "r") as f:
    data = json.load(f)
```

```python
from collections import defaultdict
move_len = defaultdict(int)
for name, pkmn in data.items():
    move_len[len(pkmn["moves"])] += 1
```

```python
for pkmn in pkmn_set:
    if pkmn not in data:
        print(pkmn)
```

```python
total_len = 0
for pkmn in data:
    total_len += len(data[pkmn]["moves"])
total_len / len(data)
```

```python
all_moves = set()
for pkmn, pkmn_data in data.items():
    all_moves.update(pkmn_data["moves"].keys())
```

```python
from poke_env.environment.move import Move

clean_move = lambda move: move.replace(" ", "").replace("-", "").lower()
```

```python
for move_str in all_moves:
    move = Move(clean_move(move_str), 1)
    if len(move.secondary) == 1:
        print(move_str, move.secondary)
    # if move.status:
    #     print(move_str, move.status)
    # if move.volatile_status:
    #     print(move_str, move.volatile_status)
```

```python
max_moves = 0
for pkmn, pkmn_data in data.items():
    max_moves = max(max_moves, len(pkmn_data["moves"]))
max_moves
```

```python
for pkmn, pkmn_data in data.items():
    if len(pkmn_data["moves"]) > 9:
        print(pkmn, pkmn_data)
```

```python
clean_name = (
    lambda move: move.replace(" ", "")
    .replace("-", "")
    .replace(".", "")
    .replace("’", "")
    .lower()
)
```

```python
data_cleaned = {}
for pkmn, pkmn_data in data.items():
    data_cleaned[clean_name(pkmn)] = {clean_name(move): pkmn_data["moves"][move]  for move in pkmn_data["moves"]}
```

```python
data_cleaned
```

```python
with open("gen1randbats_cleaned.json", "w") as f:
    json.dump(data_cleaned, f)
```
